```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd
```

```python
planilha = pd.read_csv('/home/vant3d/Documents/Zed2/SvoDisparity/PlanilhaErro.csv')

planilha
```

```python
erro_performance = planilha['Erro_performance'].values.tolist()

for i in range(0, len(erro_performance)):
    erro_performance[i] = abs(float(erro_performance[i].replace(',', '.')))

erro_performance.sort()
erro_performance

errosMarcacoesPerformance = dict(map(lambda i, j: (i, j) , planilha.index, erro_performance))

errosPERFORMANCE = pd.DataFrame(errosMarcacoesPerformance.values(), errosMarcacoesPerformance.keys(), columns= ['|Erros|'])

plt.scatter(planilha['Medicao Real'], erro_performance)
plt.ylabel('Erro')
plt.xlabel('Marcações')
plt.title('Erros PERFORMANCE')
plt.show()

errosPERFORMANCE
```
