```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd
```

```python
planilha = pd.read_csv('/home/vant3d/Downloads/ZED2 Medicoes - Página1.csv')
planilha.rename(columns = {'Unnamed: 0':'Marcações'}, inplace = True)
planilha.pop('Unnamed: 7')
planilha.pop('Unnamed: 5')

planilha
```

```python
marcaçoes = planilha['Marcações'].values.tolist()
medicaoReal = planilha['Medicao Real'].values.tolist()

medicaoMarcacoes = dict(map(lambda i, j: (i, j) , marcaçoes, medicaoReal))
medicaoMarcacoes
```

```python
erro_neural = planilha['Erro_neural'].values.tolist()

for i in range(0, len(erro_neural)):
    erro_neural[i] = abs(float(erro_neural[i].replace(',', '.')))

erro_neural.sort()
erro_neural

errosMarcacoesNeural = dict(map(lambda i, j: (i, j) , marcaçoes, erro_neural))

errosNEURAL = pd.DataFrame(errosMarcacoesNeural.values(), errosMarcacoesNeural.keys(), columns= ['|Erros|'])

plt.scatter(medicaoReal, erro_neural)
plt.ylabel('Erro')
plt.xlabel('Marcações')
plt.title('Erros NEURAL')
plt.show()

errosNEURAL
```

```python
erro_ultra = planilha['Erro_ultra'].values.tolist()

for i in range(0, len(erro_ultra)):
    erro_ultra[i] = abs(float(erro_ultra[i].replace(',', '.')))

erro_ultra.sort()
erro_ultra

errosMarcacoesUltra = dict(map(lambda i, j: (i, j) , marcaçoes, erro_ultra))

errosULTRA = pd.DataFrame(errosMarcacoesUltra.values(), errosMarcacoesUltra.keys(), columns= ['|Erros|'])

plt.scatter(medicaoReal, erro_ultra)
plt.ylabel('Erro')
plt.xlabel('Marcações')
plt.title('Erros ULTRA')
plt.show()

errosULTRA
```

```python
erro_performance = planilha['Erro_performance'].values.tolist()

for i in range(0, len(erro_performance)):
    erro_performance[i] = abs(float(erro_performance[i].replace(',', '.')))

erro_performance.sort()
erro_performance

errosMarcacoesPerformance = dict(map(lambda i, j: (i, j) , marcaçoes, erro_performance))

errosPERFORMANCE = pd.DataFrame(errosMarcacoesPerformance.values(), errosMarcacoesPerformance.keys(), columns= ['|Erros|'])

plt.scatter(medicaoReal, erro_performance)
plt.ylabel('Erro')
plt.xlabel('Marcações')
plt.title('Erros PERFORMANCE')
plt.show()

errosPERFORMANCE
```

```python
erro_performance1080 = planilha['Erro_performance1080'].values.tolist()

for i in range(0, len(erro_performance1080)):
    erro_performance1080[i] = abs(float(erro_performance1080[i].replace(',', '.')))

erro_performance1080.sort()
erro_performance1080

errosMarcacoesPerformance1080 = dict(map(lambda i, j: (i, j) , marcaçoes, erro_performance1080))

errosPERFORMANCE1080 = pd.DataFrame(errosMarcacoesPerformance1080.values(), errosMarcacoesPerformance1080.keys(), columns= ['|Erros|'])

plt.scatter(medicaoReal, erro_performance1080)
plt.ylabel('Erro')
plt.xlabel('Marcações')
plt.title('Erros PERFORMANCE1080')
plt.show()

errosPERFORMANCE1080
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(111)

ax1.scatter(medicaoReal, erro_neural, s= 20, c='b', marker='s', label='NEURAL')
ax1.scatter(medicaoReal, erro_ultra, s= 20, c='r', marker='o', label='ULTRA')
ax1.scatter(medicaoReal, erro_performance, s= 20, c='g', marker='v', label='PERFORMANCE')
ax1.scatter(medicaoReal, erro_performance1080, s= 20, c='y', marker='p', label='PERFORMANCE1080')
plt.legend(loc='upper left')
plt.show()
```
